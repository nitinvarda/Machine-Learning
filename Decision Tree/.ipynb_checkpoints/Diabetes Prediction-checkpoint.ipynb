```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
df = pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
df.head()
```

```python
#independent and dependent features
X = df
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train
```

```python
X_train.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
```

## Hyperparameter tuning

```python
param = {
    'criterion':["squared_error", "friedman_mse", "absolute_error"],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':["auto","sqrt","log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(regressor,param_grid=param,cv=5,scoring='neg_mean_absolute_error')
```

```python
# import warnings
# warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
selectedModel = DecisionTreeRegressor(criterion='friedman_mse', max_depth=3, max_features='log2', splitter='best')
selectedModel.fit(X_train,y_train)
```

```python
y_selected_pred = selectedModel.predict(X_test)
```
